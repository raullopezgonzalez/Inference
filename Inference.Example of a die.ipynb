```python
import random 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N = 10000 
trials = range(1,N+1)
pheads = []; nheads = 0

for i in trials:
    flip = random.randint(0, 1)
    nheads += flip
    pheads.append(nheads/i)
    
plt.plot(trials,pheads)
plt.xlabel('# of trials')
plt.ylabel('Relative freq')
plt.show()
```

```python
N = 10000 
trials = range(1,N+1)
p_ones = []; n_ones = 0

for i in trials:
    flip = random.randint(1, 6)
    if flip == 1:
        n_ones += flip
        p_ones.append(n_ones/i)
    else:
        n_ones += 0
        p_ones.append(n_ones/i)

plt.plot(trials,p_ones)
plt.xlabel('# of trials')
plt.ylabel('Relative freq')
plt.show()
```
